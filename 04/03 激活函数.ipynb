## Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_grad(x):
    return (1 - sigmoid(x)) * sigmoid(x)
```

```python
def step_fun(x):
    return np.array(x>0, dtype='i4')

x = np.arange(-6,6,0.1)

plt.plot(x,sigmoid(x))  # 蓝线
plt.plot(x,sigmoid_grad(x),'y*')  # 黄色
plt.plot(x,step_fun(x),'k--')  # 黑色虚线
plt.show()
```

## ReLU
h(x) = 0  x<=0  
h(x) = x  x>0  

```python
def relu(x):
    return np.maximum(0,x)

def relu_grad(x):
    return np.where(x>0,1,0)
```

```python
x = np.arange(-2,2,0.1)

plt.plot(x,relu(x))  # 蓝线
plt.plot(x,relu_grad(x),'y*')  # 黄色
plt.show()
```
