```python
import numpy as np

def f(x):
    return 1/20 * x[0]**2 + x[1]**2
```

```python
arr = np.array([2,3])
f(arr)
```

```python
# 画出f(x)的三维图
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.arange(-100,100,1)
y = x.copy()
x,y=np.meshgrid(x, y)  #x,y都是二维数组 (100,100)

X = np.array([x,y])  # X 是三维数组 (2,100,100)
z = f(X) # z 是二维数组 (100,100)

fig = plt.figure(figsize=(8,10))
ax = fig.gca(projection='3d')
ax.plot_surface(X[0],X[1],z) # 这里x,y轴不一定要一维数组，二维也行，只要数据一一对应就可

ax.set_xlabel('x',size=15)
ax.set_ylabel('y',size=15)
ax.set_zlabel('z',size=15)
```

```python
# f(x)的等高线图
x = np.arange(-100,100,1)
y = x.copy()
x,y=np.meshgrid(x, y)

X = np.array([x,y])
z = f(X)

fig = plt.figure(figsize=(8,6))
ax = fig.gca()
ax.contour(X[0],X[1],z,levels=30)  # 这里x,y轴不一定要一维数组，二维也行，只要数据一一对应就可

ax.set_xlabel('x',size=20)
ax.set_ylabel('y',size=20)
```
